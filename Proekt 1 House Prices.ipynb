```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
```

```python
# Load the California Housing dataset
california = fetch_california_housing()
x=california.data
y=california.target
```

```python
feature_names=california.feature_names
target_names=california.target_names
# print(feature_names)
# print(target_names)
```

```python
data = fetch_california_housing(as_frame=True)
df = data.frame
# print(df)
```

```python
# Preprocess the data
# Handle missing values
x_broj_na_sobi=df["AveRooms"].mean()
df["AveRooms"].fillna(x_broj_na_sobi, inplace=True)
x_broj_spalni_sobi=df["AveBedrms"].mean()
df["AveBedrms"].fillna(x_broj_spalni_sobi, inplace=True)
y_prosecna_cena=df["MedHouseVal"].mean()
df["MedHouseVal"].fillna(y_prosecna_cena, inplace=True)
```

```python
# Split the data into training and testing sets
A=df[['AveRooms','AveBedrms']]
B=df['MedHouseVal']
print(A, B)
```

```python
# Split the data into training and testing sets
A_train, A_test, B_train, B_test=train_test_split(A, B, test_size=0.33, random_state=42)
```

```python
# Build the linear regression model
model=LinearRegression()
```

```python
# Train the model and evaluate it   # Fit the model to the training data   # Make predictions on the training and testing data
model.fit(A_train, B_train)
train_predictions=model.predict(A_train)
test_predictions=model.predict(A_test)
print("Train:", train_predictions)
print("Test:", test_predictions)
```

```python
# Scatter plot for training set
plt.scatter(A_train['AveRooms'], B_train, color='blue', label='Actual')
plt.scatter(A_train['AveRooms'], train_predictions, color='red', label='Predicted')
plt.xlabel('AveRooms')
plt.ylabel('MedHouseVal')
plt.title('Actual vs Predicted (Training Set)')
plt.legend()
plt.show()
```

```python
plt.scatter(A_train['AveBedrms'], B_train, color='green', label='Actual')
plt.scatter(A_train['AveBedrms'], train_predictions, color='yellow', label='Predicted')
plt.xlabel('AveBedrms')
plt.ylabel('MedHouseVal')
plt.title('Actual vs Predicted (Training Set)')
plt.legend()
plt.show()
```

```python
# Scatter plot for testing set
plt.scatter(A_test['AveRooms'], B_test, color='blue', label='Actual')
plt.scatter(A_test['AveRooms'], test_predictions, color='red', label='Predicted')
plt.xlabel('AveRooms')
plt.ylabel('MedHouseVal')
plt.title('Actual vs Predicted (Testing Set)')
plt.legend()
plt.show()
```

```python
plt.scatter(A_test['AveBedrms'], B_test, color='green', label='Actual')
plt.scatter(A_test['AveBedrms'], test_predictions, color='yellow', label='Predicted')
plt.xlabel('AveBedrms')
plt.ylabel('MedHouseVal')
plt.title('Actual vs Predicted (Testing Set)')
plt.legend()
plt.show()
```
